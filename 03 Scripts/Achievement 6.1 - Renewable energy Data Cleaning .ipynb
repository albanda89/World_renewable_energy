```python
#create a function to find outliers using IQR

def find_outliers_IQR(df_RenE_C):

   q1=df_RenE_C.quantile(0.25)

   q3=df_RenE_C.quantile(0.75)

   IQR=q3-q1

   outliers = df_RenE_C[((df_RenE_C<(q1-1.5*IQR)) | (df_RenE_C>(q3+1.5*IQR)))]

   return outliers
```

```python
outliers = find_outliers_IQR(df_RenE_C['Access_Electracity(%)-Rural_P'])

print('number of outliers: '+ str(len(outliers)))

print('max outlier value: '+ str(outliers.max()))

print('min outlier value: '+ str(outliers.min()))

outliers
```

```python
outliers = find_outliers_IQR(df_RenE_C['Access_Electracity(%)-Urban_P'])

print('number of outliers: '+ str(len(outliers)))

print('max outlier value: '+ str(outliers.max()))

print('min outlier value: '+ str(outliers.min()))

outliers
```
