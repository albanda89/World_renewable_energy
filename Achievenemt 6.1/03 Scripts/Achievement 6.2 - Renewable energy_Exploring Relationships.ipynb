#### Create a correlation heatmap using seaborn:( Using already cleaned data set ) - 

```python
# identify the variables which suitable for exploratory visual analysis in the data set and preparing subset with necessary varibles and removing the rest of the varibles.

df_ReneE_C2.columns
```

```python
# Reorder dataframe columns from a list of column names

cols= ['Country Name', 'Country Code', 'Time', 'Time Code','Access_Electracity(%)-Rural_P','Access_Electracity(%)-Urban_P',
       'Access_Electracity(%)-Total_P','RenewableEOutput(gwh)', 'Total_Electracity_output(gwh)','RenewableE/TotalE(%)',
       'RenewableE_consumption(TJ)','Total_final_E_consumption(TFEC)', 'Renewable_E/TFEC(%)']

df_ReneE_C3 = df_ReneE_C2[cols]

df_ReneE_C3.head(5)
```

```python
# Create a subset excluding the unnecessary columns :

sub = df_ReneE_C3[['Country Name', 'Time', 'Access_Electracity(%)-Rural_P','Access_Electracity(%)-Urban_P',
       'Access_Electracity(%)-Total_P','RenewableEOutput(gwh)', 'Total_Electracity_output(gwh)','RenewableE/TotalE(%)',
       'RenewableE_consumption(TJ)','Total_final_E_consumption(TFEC)', 'Renewable_E/TFEC(%)']]
```

```python
sub.head(5)
```

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(8,8))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr = sns.heatmap(sub.corr(), annot = True, ax = ax)

# The `annot` argument allows the plot to place the correlation coefficients onto the heatmap.
```

### Scatterplots:

```python
# Create a scatterplot for the "Access_Electracity(%)-Total_P" and "RenewableEOutput(gwh)" columns in seaborn

sns.lmplot(x = 'Access_Electracity(%)-Total_P', y = 'RenewableEOutput(gwh)', data =df_ReneE_C3)
plt.figure(figsize=(10,10))
```

## Categorical Plots:

```python
# Use a histogram to visualize the distribution of the variables. 

sns.histplot(df_ReneE_C3['Access_Electracity(%)-Total_P'], bins = 10, kde = True)
```

```python
df_ReneE_C3.loc[df_ReneE_C3['Access_Electracity(%)-Total_P'] >0 & (df_ReneE_C3['Access_Electracity(%)-Total_P'] <=40), 'Access category'] = 'low AE'
```
